```python
#imports
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

**HTTP Request**

```python
#HTTP Request
#store website in variable
website = 'https://www.automobile.tn/fr/occasion/1'
#Get Request
response = requests.get(website)
#Status Code
response.status_code
```

**Soup Object**

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

**Results**

```python
results = soup.find_all('div', {'class' : 'inner'})
```

```python
len(results)
```

```python
results[0]
```


**Target necessary data**

```python
# Name
# Mileage
# Localite
# Year
# Energie
# Boite
# Price
# Modèle de la voiture
Name = soup.find('h2').get_text(strip=True)

# Kilométrage
mileage = soup.find('li', {'class': 'road'}).get_text(strip=True)

# Année
year = soup.find('li', {'class': 'year'}).get_text(strip=True)

# Boîte de vitesses
Boite = soup.find('li', {'class': 'boite'}).get_text(strip=True)

# Carburant
fuel = soup.find('li', {'class': 'fuel'}).get_text(strip=True)

# Transmission
transmission = soup.find('li', {'class': 'transmission'}).get_text(strip=True)

# Puissance fiscale
horsepower = soup.find('li', {'class': 'horsepower'}).get_text(strip=True)

# Description
#description = soup.find('div', {'class': 'description'}).find('p').get_text(strip=True)

# Localisation
localite = soup.find('span', {'class': 'fw-medium text-muted d-inline-flex align-items-center'}).get_text(strip=True)

# Prix
price = soup.find('div', {'class': 'price'}).get_text(strip=True).replace('DT', '').strip()

# Affichage des résultats
print(f"Modèle: {Name}")
print(f"Kilométrage: {mileage}")
print(f"Année: {year}")
print(f"Boîte de vitesses: {Boite}")
print(f"Carburant: {fuel}")
print(f"Transmission: {transmission}")
print(f"Puissance fiscale: {horsepower}")
#print(f"Description: {description}")
print(f"Localisation: {localite}")
print(f"Prix: {price} DT")
```

```python
#Put everything together inside a For-Loop

name = []
mileage = []
localite = []
year = []
energie = []
boite = []
price = []
transmission = []
horsepower = []

for result in results:

    # name
    try:
        name.append(result.find('h2').get_text())
    except:
        name.append('n/a')

    # mileage
    try:
        mileage.append((result.find('li', {'class':'road'}).get_text()).replace(' ','').strip().replace('\n', ''))
    except:
        mileage.append('n/a')

    # localite
    try:
        localite.append(result.find('span', {'class':'fw-medium text-muted d-inline-flex align-items-center'}).get_text().strip().replace('\n', ''))
    except:
        localite.append('n/a')

    # year
    try:
        year.append(result.find('li', {'class':'year'}).get_text().strip().replace('\n', ''))
    except:
        year.append('n/a')

    # energie
    try:
        energie.append(result.find('li', {'class':'fuel'}).get_text().strip().replace('\n', ''))
    except:
        energie.append('n/a')

    # Transmission
    try:
        transmission.append(result.find('li', {'class': 'transmission'}).get_text().strip().replace('\n', ''))
    except:
        transmission.append('n/a')
# Puissance fiscale
    try:
        horsepower.append(result.find('li', {'class': 'horsepower'}).get_text().strip().replace('\n', ''))
    except:
        horsepower.append('n/a')
    # boite
    try:
        boite.append(result.find('li', {'class':'boite'}).get_text().strip().replace('\n', ''))
    except:
        boite.append('n/a')

    #price
    try:
        price.append(((result.find('div', {'class':'price'}).get_text()).replace(' ','').strip('\n')).strip('DT'))
    except:
        price.append('n/a')
```

**Create Pandas Dataframe**

```python
# dictionary
car_dealer = pd.DataFrame({'Name': name, 'Mileage':mileage, 'Localite':localite,
                           'Year': year, 'Transmission': transmission, 'Horsepower': horsepower, 'Energie': energie, 'Boite': boite, 'Price': price})
```

**Output in Excel**

```python
car_dealer.to_excel('cars_single_page.xlsx', index=False)
```

**Pagination**  (all pages)

```python
name = []
mileage = []
localite = []
year = []
energie = []
boite = []
price = []
transmission = []
horsepower = []

for i in range (1,50):

    # website in variable
    website = 'https://www.automobile.tn/fr/occasion/' + str(i)

    # request to website
    response = requests.get(website)

    # soup object
    soup = BeautifulSoup(response.content, 'html.parser')

    # results
    results = soup.find_all('div', {'class' : 'inner'})


    # loop through results
    for result in results:
      # name
      try:
        name.append(result.find('h2').get_text())
      except:
        name.append('n/a')

    # mileage
      try:
        mileage.append((result.find('li', {'class':'road'}).get_text()).replace(' ','').strip().replace('\n', ''))
      except:
        mileage.append('n/a')

    # localite
      try:
        localite.append(result.find('span', {'class':'fw-medium text-muted d-inline-flex align-items-center'}).get_text().strip().replace('\n', ''))
      except:
        localite.append('n/a')

    # year
      try:
        year.append(result.find('li', {'class':'year'}).get_text().strip().replace('\n', ''))
      except:
        year.append('n/a')

    # energie
      try:
        energie.append(result.find('li', {'class':'fuel'}).get_text().strip().replace('\n', ''))
      except:
        energie.append('n/a')

    # Transmission
      try:
        transmission.append(result.find('li', {'class': 'transmission'}).get_text().strip().replace('\n', ''))
      except:
        transmission.append('n/a')
# Puissance fiscale
      try:
        horsepower.append(result.find('li', {'class': 'horsepower'}).get_text().strip().replace('\n', ''))
      except:
        horsepower.append('n/a')

    # boite
      try:
        boite.append(result.find('li', {'class':'boite'}).get_text().strip().replace('\n', ''))
      except:
        boite.append('n/a')

    #price
      try:
        price.append(((result.find('div', {'class':'price'}).get_text()).replace(' ','').strip('\n')).strip('DT'))
      except:
        price.append('n/a')
```

```python
# dictionary
car_dealer = pd.DataFrame({'Name': name, 'Mileage':mileage, 'Localite':localite,
                             'Year': year,'Transmission': transmission, 'Horsepower': horsepower, 'Energie': energie, 'Boite': boite, 'Price': price})
```
